```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.colors as mcolor
```

```python
df2 = pd.read_csv("amazon.csv")
df2.head()
```

```python
df = df2.copy()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head(20)
```

```python
df.drop_duplicates("product_id")
```

```python
df["category"].head()
```

```python
df_catg_split = df["category"].str.split("|", expand = True)
df_catg_split.rename(columns={
     0: "category1",
     1: "category2",
     2: "category3",
     3: "category4",
     4: "category5",
     5: "category6",
     6: "category7",
     7: "category8"
 }, inplace= True)
```

```python
category_1 = []
for i in df["category"]:
    for k in i.split("|"):
        category_1.append(k)
    
all_categories = pd.Series(category_1).value_counts()
all_categories
```

```python
df= pd.concat([df,df_catg_split], axis=1)
df.head()
```

```python
df.drop(columns=["category", "review_content","img_link","product_link"], inplace = True)
```

```python
df["rating"].value_counts()
```

```python
#df.loc[df["rating"].isin(["|"])]= "0"
df["rating"]= df["rating"].replace("|","0")
```

```python
df["rating"].value_counts()
```

```python
#columns need to be changed
df.loc[df["rating"].isin(["|"])] = "0"

columns_change = ["discounted_price","actual_price","discount_percentage","rating_count","rating"]

for i in columns_change:
    df[i]=df[i].str.replace("₹","")
    df[i]=df[i].str.replace("%","")
    df[i]=df[i].str.replace(",","")
    df[i]= df[i].astype("float")
    
    print(f"finished >> {i}")
df.head()
    
```

```python
df["product_name"].nunique()
```

```python
# product(top 15) rated by most people
most_rated_prod = df.groupby("product_name")["rating_count"].sum().sort_values(ascending= False)[:15]
plt.barh(most_rated_prod.index,most_rated_prod.values, color=["pink"], edgecolor="black")
plt.title("product rated by most people")
plt.xticks(rotation=45)
for i, v in enumerate(most_rated_prod):
    plt.text(v+0.5, i, str(v), fontsize=10)
```

```python
df["discount_percentage"].sort_values(ascending= False)
```

```python
# most dixcounted product(top15)
most_disc_prod = df.groupby("discount_percentage")["product_name"].nunique()[:15]
most_disc_prod = most_disc_prod.sort_values(ascending= False)
cmap1= mcolor.LinearSegmentedColormap.from_list("mycmap",['#00d0fa','#000000'])
color1= cmap1(np.linspace(0,1,len(most_disc_prod.index)))
wedgeprops ={'linewidth':0.7,'edgecolor':'black'}
plt.pie(most_disc_prod, labels=most_disc_prod.values, autopct="%0.2f%%",wedgeprops=wedgeprops)
plt.show()
```

```python
#data =df[df["discounted_price"]>20000]
most_disc_catg = df.groupby("category1")["discounted_price"].sum().sort_values(ascending= False)
cmap2= mcolor.LinearSegmentedColormap.from_list("mycmap",['pink','black'])
color2 = cmap2(np.linspace(0,0.75,len(most_disc_catg.index)))
plt.bar(most_disc_catg.index,most_disc_catg.values, color=color2, edgecolor='black')
plt.xticks(rotation=45)
for i in range(len(most_disc_catg.index)):
    plt.text(i, most_disc_catg.values[i],round(most_disc_catg[i],1),ha='center',fontsize=10)
```
